# Pandas DataReader FRED Function

```python
# gather data from FRED

def gather_data_tb(data_code, 
                start, 
                end = datetime.datetime.today(), 
                freq = "M"):
    i = 0
    # dct.items() calls key and value that key points to
    for key, val in data_code.items():
        if i == 0:
            # Create dataframe for first variable, then rename column
            df = pdr.DataReader(
                val, "fred", start, end).resample(freq).mean()
            df.rename(columns = {val:key}, inplace = True) 
            # setting i to None will cause the next block of code to execute,
            # placing data within df instead of creating a new dataframe for
            # each variable
            i = None
        else:
            # If dataframe already exists, add new column
            df[key] = pdr.DataReader(val, "fred", start, end).resample(freq).mean()

    return df
```

# What is currency in circulation?
"According to investopedia it is, "Currency in circulation is all of the money that has been issued by a country's monetary authority, minus cash that has been removed from the system... Currency in circulation can also be thought of as currency in hand because it is the money used throughout a country's economy to buy goods and services. "

```python
data_codes = {"Base: Total":"BOGMBASEW",
              "Base: Currency in Circulation": "WCURCIR"}
```

```python
def gather_data(data_codes,start,end=datetime.datetime.today(),
                freq='M'):
    df_init = False
    for key, code in data_codes.items():
        if df_init == False:
            df = pdr.DataReader(
            code, "fred", start, end).resample(freq).mean() #.first() or .last()
            df.rename(columns = {code:key}, inplace=True)
            df_init = True
        else:
            df[key] = pdr.DataReader(
            code, "fred", start, end).resample(freq).mean()
    return df
```

# Function to plot lines

```python
from matplotlib.backends.backend_pdf import PdfPages
import os
```

```python
def plot_lines(df,
               linewidth=5,
               figsize=(40,20),
               secondary_y = None,
               legend = True,
               pp = None,
               save_fig = False,
               global_fontsize=32):
    plt.rcParams.update({"font.size":global_fontsize})
    fig, ax = plt.subplots(figsize = figsize)
    df.dropna(axis = 0, how = 'all').plot.line(linewidth=linewidth,
                                               ax=ax, 
                                               secondary_y=secondary_y, 
                                               legend=legend)
    ax.set_xlabel('') # Remove 'Date' label
    ax.tick_params(axis = 'x', rotation = 90) # Rotate tick labels 90 degrees
    ax.tick_params(axis = 'both', length = 0, which = 'both') # Remove tick labels on both axes
    vals = ax.get_yticks() # Assign yticks to vals
#     ax.set_yticklabels([round(x,2) for x in vals]) # Round values in vals and set to yticklabels
    ax.set_yticklabels([str(y) + "%" for y in vals])
    
    # save figure as file
    remove_chars = "[]:$'\\'"
    filename=str(list(df.keys()))
    #remove chars from filename
    for char in remove_chars:
        filename=filename.replace(char,"")
    if save_fig:
        folder = "Monetary Plots"
        try:
            os.mkdir(folder)
        except:
            pass
        plt.savefig(folder + "/" + filename[:50] + "line.png",
                   bbox_inches='tight')
            
    return fig, ax
    
plot_data = data[['Effective Federal Funds Rate',
                  'Interest on Reserve Balances']]

fig,ax=plot_lines(plot_data, save_fig = True)

ax.set_title("Fred Data", color = 'olivedrab')
```

#### What is excess reserves? These are the funds held at the fed which exceed the depository reserve requirement imposed by the fed. 

```python
def plot_stacked(df,
                stack_vars,
                linewidth=2.5,
                figsize=(40,20),
                total_var=False,
                global_fontsize = 40):
    plt.rcParams.update({"font.size":global_fontsize})
    plt.rcParams['axes.ymargin']=0.01
    plt.rcParams['axes.xmargin']=0.01
    # Make the subplots
    fig, ax = plt.subplots(figsize=figsize)
    # Plot first to get the full picture
    if total_var != False:
        df[total_var].plot.line(linewidth=linewidth,ax=ax, c = 'k', label=total_var,ls='--')

    area_df = df[stack_vars]
    area_df.plot.area(stacked=True,
                     linewidth=0,
                     ax=ax)
    ax.legend(loc=2, ncol=2)

    return fig, ax
stack_vars = money_vars[1:]
total_var = money_vars[0]

fig,ax = plot_stacked(data,stack_vars=stack_vars,total_var=total_var)
```

# Interbank/Overnight Lending Market
- This is where the central bank is lending to outher banks to support the fractional reserve banking model. The rate set here is typically used as the benchmark for the overall lending market. When banks aren't able to service their reserve requirements, they borrow from other banks which have too much. This rate is also important in controlling overall lending activity due to the influence higher or lower interest rates have on business and consumer lending.

- The point of repo operations is to control the fed funds rate as set by the FOMC.

from bankrate.com "when the Fed sells a security to a counterparty and then agrees to buy back that security, itâ€™s a transaction known as a â€œreverse repo.â€

also from bankrate.com repo = "One firm sells securities to a second institution and agrees to purchase back those assets for a higher price by a certain date, typically overnight"

- Please ignore the % signs.
