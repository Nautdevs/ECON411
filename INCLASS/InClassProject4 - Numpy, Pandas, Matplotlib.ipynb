### Ingest the data

```python
incdat = pd.read_excel("https://www2.census.gov/programs-surveys/cps/tables/hinc-06/2019/hinc06.xls", header=[7])
# Rename first column to "Household Income"
incdat = incdat.rename(columns = {"Unnamed: 0": "Household Income"})
```

```python
# There was an issue with the list(range()) function. This works.
income_levels = [i for i in range(0, 201, 5)]
income_levels.append(250)
income_levels = [str(i) + "k" for i in income_levels]
income_levels.append("250k+")
print(income_levels)
```

# Goal: calculate percentage of people in bracket, then plot that against the cumulate percentage up to that bracket.

```python
# Pass the array of income levels as the values for the Income Level column.
income_data = pd.DataFrame({"Income Level":np.array(income_levels),
# Pass income to the Householders in Bracket column.
                           "Households in Bracket":income})


# Use .cumsum() method on the column to get a running cumulative sum.
# .cumprod() works as well
income_data["Cumulative"] = income_data["Households in Bracket"].cumsum()

# Set income level column as the index column. The inplace means do it to the same index
income_data.set_index("Income Level", inplace = True)

#Display the data frame
income_data
```

```python
#income_data.plot.line(secondary_y = "Cumulative")
```

```python
# Get total households
totalHouses = income_data["Households in Bracket"].sum().round(2)
```

```python
# Calculate percentage in each bracket. The bracket # households / total households then convert from decimal to whole  and round.
income_data["% Housesholds in Bracket"] = income_data["Households in Bracket"].div(totalHouses).mul(100)
```

```python
income_data["Cumulative %"] = income_data["% Housesholds in Bracket"].cumsum().round(2)
```

```python
income_data
```

```python
# rcParams allows you to access global variables in mpl and set a default value for them. (Changes font size for everything, unless explicated otherwise)
plt.rcParams.update({"font.size" : 16})
plot_df = income_data[["% Housesholds in Bracket", "Cumulative %"]]


fig, ax = plt.subplots(figsize = (15, 7))

# Use secondary y for second axes
plot_df.plot.bar(secondary_y = "Cumulative %", 
                 ax = ax,
                 #Bar width
                width = 1,
                 # Opacity
                alpha = .6,
                )

# Generate new labels
genXlabels = ["$" + str(i) for i in plot_df.index]
# Set labels
ax.set_xticklabels(genXlabels)

#Same as above
genY1labels = [str(y) + "%" for y in ax.get_yticks()]
ax.set_yticklabels(genY1labels)
```

```python
#Select only every 5 bins and assign them to a new bin in a new dataframe
_25k_bin = income_data.loc[::5]
# Select the 250k+ bin and assign it to the final bin
_25k_bin.loc["250k+"] = income_data.iloc[-1]

# Assign difference between cumulative values to the Households in Bracket numbers
#Also .fillna(0) set the null values to zero
_25k_bin["Households in Bracket"] =_25k_bin["Cumulative"].diff().fillna(0)

#Calculate cumulative because the prior was incorrect
_25k_bin["% Houses in Bracket"] = _25k_bin["Households in Bracket"].div(
    totalHouses).mul(100)

del _25k_bin["% Housesholds in Bracket"]
_25k_bin
```
