# Introduction
    Investors seek out optimal asset allocation strategies which maximize their returns and minimize their risk. One of the more popular approaches to achieving this is through quantitative computational models which utilize Modern Portfolio Theory. This theory provides a method for construction a diversified portfolio which balance risk and return through the selection of assets with differing levels of risk and potential for returns. 
    
    In this paper, I introduce a computational model which aims to generate an optimal asset allocation using the aforementioned methods from Modern Portfolio Theory and extension thereof called the Fama-French five factor model. Modern Portfolio Theory accounts for factors like the risk-free rate, expected returns, and return risks while Fama-French's Three Factor Model takes into account additional risk related factors such as the firm's size, operating profitability, book-to-market equity, and character of investment. By taking into account these factors the computational model is better able to accomodate sources of risk in the market and result in a more robust portfolio. 

    I begin the paper by reviewing the basic principles of Modern Portfolio Theory and how it is used in practice to construct an optimal portfolio. Then I will walk readers through its construction and interpret results. Then I will review Fama-French's Five-Factor model. 

    This computational model can be used by investors to construct a portfolio which takes advantage of optimal allocation strategies that balance risk and return enabling investors to achieve their investment goals.

# Limitations
    Deterministic computational programs like this uses en masse could create a situation where everyone defaults to methods like this and thereby create a negative feedback loop within the financial markets exacerbating volatility. This would be especially true for real-time portfolio construction programs where decisions are made near instantaneously. If one program quit investing in one firm and decided to focus on another asset class or firm and another program picked up on the same data stream resulting in that decision it would result in a cascading meltdown throughout the financial markets. Although there exists restrictions such as the market being open for a set period of time for five days a week and a circuit breaker which shuts down trading in the event of extreme booms or busts.

```python
def adj_plots(df):
    level = df.columns.get_level_values(0).unique()
    # Loop through the unique values
    for col in level:
        # For each unique value, assign the original df[unique_value] to sub_df Level 1
        sub_df = df[col]
        # For each unique value within those unique values, assign to sub_cols Level 2
        sub_cols=sub_df.columns
    # Count how many columns are in level 2
    numCharts = len(sub_cols)
    
    # Level 1 plot
    # Make a big plot with subplots for each ticker
    fig, ax = plt.subplots(nrows=1, ncols=len(sub_cols), figsize = (60,8))
    
    # Loop through the # of columns in level 2
    for i in range(numCharts):
        # Call the columns data
        key = df.keys()[i]
        color = "C" + str(i)
        alpha = .35 + .15 * i
        # plot key according to position in loop
        df[key].plot.line(color=color, alpha=alpha, ax=ax[i], legend=True)
        ax[i].set_title(key, color=color, fontsize =20)
```

```python
adj_plots(concat_df)
```

# Part of #9 was figuring out how locators and formatters works in order to reduce tick labels

```python
def bar_plot(df):
    numVar = len(df.columns)
    fig, ax = plt.subplots(1, numVar, figsize = (60,8))
    for i in range(numVar):
        key = df.keys()[i]
        color = "C" + str(i+8)
        alpha = .35 + .15 * i
        df[key].plot.bar(color=color, alpha=alpha, ax=ax[i], width = .95)
        ax[i].set_title(key, color=color, fontsize =64)
        
        # Automatically chooses tick labels
        ax[i].xaxis.set_major_locator(plt.AutoLocator())
        
        #Rotes and right aligns x labels
        fig.autofmt_xdate()
```

```python
bar_plot(percent_df_2)
```
