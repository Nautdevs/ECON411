# create artificial data

```python
macroDict = {
    "GDP": {},
    "Money": {},
    "Real GDP": {},
    "Price Level": {}
}

# Select each key and assign a random value for each year.
for key in macroDict.keys():
    for i in range(1990,2069):
        macroDict[key][i] = np.random.random() * 10000
macroDict
```

```python
macroDF = pd.DataFrame(macroDict)
macroDF
```

```python
macroDF[['GDP']]
```

```python
macroDF["Velocity"] = macroDF["Price Level"].mul(macroDF["Real GDP"]).div(macroDF["Money"])

# Velocity = (Price level * Real GDP)/M
macroDF
```

```python
macroDF.iloc[0:50]
```

# Matplotlib Introduction

```python
# Remember this import
import matplotlib.pyplot as plt
```

```python
macroDF.plot()
```

```python
# This is how you plot using the method for mpl from the pd df

# Length, Width in the figsize keyword argument
fig, ax = plt.subplots(figsize = (10,10))

# Alpha makes lines more opaque.
macroDF.plot(linewidth = 2.5, alpha = .5, ax=ax, legend = False)

ax.legend(fontsize = 20)
```

### Make a chart for each key in the dataframe

```python
# Explicit is better than implicit
for key in macroDF:
    fig, ax = plt.subplots(figsize = (15,10))
    macroDF[key].plot.line(ax = ax)
    ax.set_title(key, fontsize = 20)
```

```python
macroDF.keys()[1]
```

```python
# Our goal is to create a multiplot visualization, separate plot area for each variable.
# 1. Count number of variables
numVars = len(macroDF.keys())
# 2. Create a plot with a subplot for each variable
fig, ax = plt.subplots(1, numVars, figsize = (40,5))

# To plot each variable on each plot, cycle through the index  of the keys.
for i in range(numVars):
    # Call key by index number (the list method was not working for me)
    key = list(macroDF.keys())[i]
    # set color in light of number i (C0, C1, etc)
    color = "C" + str(i)
    # 
    alpha = .35 + .15 * i
    # plot key according to index number in ax
    macroDF[key].plot.line(color = color, alpha = alpha, ax = ax[i])
    ax[i].set_title(key, color = color, fontsize = 20)
    
    
# Every attribute in a visualization should have meaning.
```
