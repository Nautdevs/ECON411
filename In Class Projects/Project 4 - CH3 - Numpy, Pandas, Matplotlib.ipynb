You can call or manipulate individual cells within a 2d array through the following:

```python
array = np.zeros((9,9))
print("Here is a 2d array of zero values", array, sep = "\n")

# Access individual cells in the 2d array and change them

array[0][1] = 1.5
array[8][2] = 3.5
array[4][2] = "Hello"
```

### Accessing individual cells within a 2d array

### Functions like the zero function include:
## .zeros_like
## .ones_like
## .empty_like

```python
threedee = [[1,2,3], [4,5,6], [7,8,9]]
anotherdee = [[9,9,9],[8,8,8]]
print(threedee)
threedee = np.array(threedee)
print("\n", threedee)
```

```python
zero = np.zeros_like(threedee)
print(zero)
anotherzero = np.zeros_like(anotherdee)
print("\n", anotherzero, '\n')

# ones_like
ones = np.ones_like(threedee)
anotherones = np.ones_like(anotherdee)
print(ones, anotherones, sep="\n\n")

# empty_like
empty = np.empty_like(anotherdee)
print("Empty_like:", empty, sep="\n") #Very strange, why does it do this? Better to use zero
```

# create artificial data

```python
macroDict = {"GDP": {},
            "Money": {},
            "Real GDP": {},
            "Price Level": {}}
for key in macroDict.keys():
    for i in range(1990,2021):
        macroDict[key][i] = np.random.random() * 10000
macroDict
```

```python
macroDF = pd.DataFrame(macroDict)
macroDF
```

```python
macroDF["Velocity"] = macroDF["Price Level"].mul(macroDF["Real GDP"]).div(macroDF["Money"]) 
print(macroDF)
```

```python
macroDF.loc[1998:2010]
```
